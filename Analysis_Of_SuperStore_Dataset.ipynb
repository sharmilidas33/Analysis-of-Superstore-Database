# ***Sales Trend over Time (Monthly, Yearly)***

```python
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'])
monthly_sales = dataset.groupby(['Order Date'], as_index=False).sum()

# Set the Order Date column as the index of the dataframe
monthly_sales = monthly_sales.set_index('Order Date')
# Resample the data into monthly intervals
monthly_sales = monthly_sales.resample('M').sum() # M for month

# Plot
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['Sales'])
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
yearly_sales = monthly_sales.resample('Y').sum()
# Plot
plt.figure(figsize=(15,5))
plt.plot(yearly_sales['Sales'])
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.title("yearly Sales Trend")
plt.show()
```

# ***Sales, Profit and Discount Relations***

```python
# Group the data by the discount and compute the average sales
discount_group = dataset.groupby(["Discount"]).mean()[["Sales"]]
# Convert the discount values to percentages
discount_group.index = discount_group.index.map(lambda x: f"{int(x * 100)}%")

ax = discount_group.plot(kind="bar")
ax.set_ylabel("Sales")
plt.show()
```

Sales are the highest, when Discount is of 50%.

```python
# Group the data by the discount and compute the average profits
discount_group = dataset.groupby(["Discount"]).mean()[["Profit"]]
# Convert the discount values to percentages
discount_group.index = discount_group.index.map(lambda x: f"{int(x * 100)}%")

ax = discount_group.plot(kind="bar")
ax.set_ylabel("Profit")
plt.show()
```

Profit is least when Discount is of 50% and highest when Discount is of 10%.

**Hence, the discount from 10% to 20% is the best solution, as they are great sellers and make good profits.**

# ***Building a Machine Learning Model to Predict the Profit***

```python
dataset.columns
```

**Feature Selection:**

```python
dataset1 = dataset.drop(columns=["Row ID","Order ID","Order Date","Ship Date","Customer ID","Customer Name","Country","City","State",
                                 "Postal Code","Product ID","Sub-Category","Product Name"])
dataset1.head()
```

**Label Encoding:**

```python
print("Ship Mode :", dataset1["Ship Mode"].nunique())
print("Segment :", dataset1["Segment"].nunique())
print("Region :", dataset1["Region"].nunique())
print("Category :", dataset1["Category"].nunique())
```

```python
#to convert data into numerical data
#to make data more machine readable
#to increase accuracy of the model
dataset1.replace({"Ship Mode" : {"Standard Class":0, "Second Class":1, "First Class":2, "Same Day":3},
                  "Segment":{"Consumer":0, "Corporate":1, "Home Office":2},
                  "Category":{"Office Supplies":0, "Furniture":1, "Technology":2},
                  "Region":{"West":0, "East":1, "Central":2, "South":3}},
                 inplace=True)
dataset1.head()
```

```python
x = dataset1.iloc[:, :-1]
print(x)
```

```python
y = dataset1.iloc[:, -1]
print(y)
```

**Re-scaling the Dataset:**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

num_vars = ["Sales","Quantity"]
x[num_vars] = scaler.fit_transform(x[num_vars])
print(x)
```

**Splitting the Dataset into Training and Testing Data:**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.25, random_state=100)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

**Creating a Machine Learning Model:**

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(random_state=0, n_estimators=117)
model.fit(x_train, y_train)
```

**Predicting the Test Data:**

```python
y_pred = model.predict(x_test)

from sklearn.metrics import r2_score
print("R2 Score=",r2_score(y_pred,y_test)*100)
```

```python
#plotting
fig = plt.figure()
plt.scatter(y_test,y_pred)
#plot heading
fig.suptitle('y_test vs y_pred', fontsize=20)
plt.xlabel('y_test', fontsize=18)
plt.ylabel('Y_pred', fontsize=16)
```
